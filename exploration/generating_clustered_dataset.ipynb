# Generating a clustered dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
plt.rcParams['figure.figsize'] = (16, 9)
```

We can use the **make_blobs()** function to generate isotropic Gaussian blobs to test our clustering method.

```python
X, y = make_blobs(n_samples=1000, n_features=4, centers=4)
```

```python
fig = plt.figure()
ax = plt.scatter(X[:, 0], X[:, 1])
```

We can clearly discern 4 group of points. It will be ideal to test our clustering algorithm.

## Testing Scikit K-Means implementation

```python
# Initializing KMeans
kmeans = KMeans(n_clusters=4)
# Fitting with inputs
kmeans = kmeans.fit(X)
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
centroids = kmeans.cluster_centers_
```

```python
fig = plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', c='red', s=500)
for n in range(y):
    plt.plot()
```

As expected, the Scikit K-Means implementation clusters the data correctly.

It somewhat proves that clusters can be formed from this generated data.

## Generating a dataset to test our implementation against

```python
X_as_dataframe = pd.DataFrame(data=X, columns=["feature_"+str(index) for index in range(1,5)])
```

```python
y_as_dataframe = pd.DataFrame(data=y, columns=["cluster"])
```

```python
clustered_data = pd.concat([X_as_dataframe, y_as_dataframe], axis=1)
```

```python
clustered_data.head().append(clustered_data.tail())
```

```python
clustered_data.to_csv(path_or_buf="./exploration/data/generated_clustered_data.csv",
                      encoding='utf-8', 
                      index=False)
```
